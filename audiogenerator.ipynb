```python
import zipfile
import os

base = "/content/drive/MyDrive/Senior_Design"

# Unzip RAVDESS
with zipfile.ZipFile(f"{base}/RAVDESS/Audio_Speech_Actors_01-24.zip", 'r') as z:
    z.extractall(f"{base}/RAVDESS/")

print("RAVDESS unzipped")

# Unzip DEMAND environments
demand_envs = ["DKITCHEN_48k", "DLIVING_48k", "PRESTO_48k", "SPSQUARE_48k", "OHALLWAY_48k", "NPARK_48k"]
for env in demand_envs:
    zip_path = f"{base}/DEMAND/{env}.zip"
    if os.path.exists(zip_path):
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(f"{base}/DEMAND/")
        print(f"{env} unzipped")
    else:
        print(f"WARNING: {zip_path} not found")
```
